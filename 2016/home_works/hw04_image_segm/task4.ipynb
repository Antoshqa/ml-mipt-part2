<h1 align="center">Task</h1> 

Рассмотрим задачу построения колажа из нескольких изображений, к примеру несколько фотографий группы людей, на каждой кто-то отвернулся или моргнул ... Вы хотите получить одну фотографию, на которой все получились хорошо. На вход поступает K изображений а на выходе вам нужно выдать матрицу размером с изображение, где в кажом пикселе будет указано из какой картинки вам нужно его взять. Вы хотите сделать фотографию, чтобы места склейки были незаметны. 

- Введите графмодель, почему вы выбрали именно такую, приложите рисунок
- Потенциалы каких порядков вы собираетесь использовать? 
- Определите потенциалы -- какой в них физичиский смысл? почему они поощряют незаметные склеивания?
- Предлажите несколько вариантов выбора потенциалов.
