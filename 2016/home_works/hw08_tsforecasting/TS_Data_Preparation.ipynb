```python
ts_ns = pd.read_csv('./data/TS_dep6_ns.csv', sep=';', decimal=',', index_col='Timestamp', parse_dates=True, dayfirst=True)
ts_es = pd.read_csv('./data/TS_dep6_es.csv', sep=';', decimal=',', index_col='Timestamp', parse_dates=True, dayfirst=True)
ts_es.head(3)
```

```python
# merge ts
ts = pd.merge(ts_ns, ts_es, how='outer', left_index=True, right_index=True)
ts.ix['2006-11-01':'2007-09-30'][sorted(ts.columns)].to_csv('./data/ts_dep6.csv', sep=';', decimal='.')
```
